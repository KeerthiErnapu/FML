```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))
x=np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('sigmoid functn')
plt.show()
```

```python
def tanh(x):
  return np.tanh(x)
x=np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('tanh functn')
plt.show()
```
