```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relu(x):
  x1=[]
  for i in x:
    if i<0:
      x1.append(0)
    else:
      x1.append(i)
  return x1
x=np.linspace(-10,10)
plt.plot(x,relu(x))
plt.axis('tight')
plt.title('relu function')
plt.show()
```

```python
def softmax(x):
  return np.exp(x)/np.sum(np.exp(x),axis=0)
x=np.linspace(-10,10)
plt.plot(x,relu(x))
plt.axis('tight')
plt.title('Softmax function')
plt.show()
```
